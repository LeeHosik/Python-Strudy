# requests

응답코드

1xx : 요청을 받았고 작업 진행중 (시간이 많이 걸림)   
2xx : 사용자의 요청이 성공적으로 수행됨    
3xx : 요청은 완료되었으나 reDirection이 필요함     
4xx : 사용자의 요청이 잘못됨  
5xx : 서버의 오류가 발생함   

```python
import requests as req

URL = "https://www.google.com"
response = req.get(URL)
print(response.status_code)
```

```python
URL = "https://www.google.com"
response = req.get(URL)
print(response.status_code)
print(response.text)
```

```python
URL = "https://search.naver.com/search.naver"
params2 = {'query':'python'}
response = req.get(URL, params=params2)
print(response.status_code)
print(response.text)
```

```python
URL = "https://comic.naver.com/webtoon/detail.nhn"
params2 = {'titleID':'687915','no':'157'}
response = req.get(URL, params=params2)
print(response.status_code)
print(response.text)
```

```python
URL = "https://raw.githubusercontent.com/naver/naver-openapi-guide/draft/naver-openapi-swagger.json"
response = req.get(URL)
result = response.json()
print(response.status_code)
print(type(result))
print(result)
```
