```python
import requests
from bs4 import BeautifulSoup
```

```python
req = requests.get('http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108')
html = req.text
soup = BeautifulSoup(html,'html.parser')
```

```python
for i in soup.find_all("location"):
    print(i.find('city'))
```

```python
# 서울로 test

seoul = soup.find('location')
# tmef : 날짜, wf: 날씨, tmn: 최저기온, tmx: 최고기온

for i in seoul.find_all('data'):
    print(i.find('tmef').text) # 하루에 0,12시 기준으로 2번 나옴 
    print(i.find('wf').text)
    print('최저기온 :',i.find('tmn').text)
    print('최고기온 :',i.find('tmx').text)
    print('-----------------------------')
```

```python
# 도시별 날씨 데이터 구축
weather = {}

for i in soup.find_all('location'):
    weather[i.find('city').text] = []
    for j in i.find_all("data"):
        temp=[]
        temp.append(j.find('tmef').text)
        temp.append(j.find('wf').text)
        temp.append(j.find('tmn').text)
        temp.append(j.find('tmx').text)
        weather[i.find('city').text].append(temp)
```

```python
print(weather['서울'])
```

```python
# save csv file
# 현재 dictionary로 되어 있는데 List형태로 변환 필요 

#test
for i in weather:
    for j in weather[i]:
        print(i, j[0], j[1],j[2], j[3])
    print("-"*30)
```

```python
# List 로 생성하기
tempList = []
for i in weather:
    for j in weather[i]:
        tempList.append((i, j[0], j[1],j[2], j[3]))

tempList
```

```python
import pandas as pd
```

```python
weatherDataFrame = pd.DataFrame(tempList, columns=['지역','날짜','날씨','최저온도','최고온도'])
weatherDataFrame.head()
```

```python
# save File
weatherDataFrame.to_csv('../Data/weatherList.csv', index=False)
```
