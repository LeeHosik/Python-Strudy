```python
from bs4 import BeautifulSoup
from urllib import parse
import urllib.request as req
import time
import pandas as pd
```

```python
url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver"
res = req.urlopen(url)
# BeautifulSoup으로 분석하기
soup = BeautifulSoup(res, "html.parser")
```

#old_content > table > tbody > tr:nth-child(25) > td.title > div

```python
movies = soup.select(".tit3")
movie=[]
RankLists=[]
count = 1

for mo in movies:
    movie.append(mo.text.strip())
    RankLists.append(count)
    count += 1
    
print(movie)
print(len(movie))
print(RankLists)
```

```python
# countLists 만으로 DataFrame 생성
df = pd.DataFrame(data=RankLists, columns=['Index']) #< columns는 [] 로 작성해야댐 
df['Title'] = pd.Series(movie)
df
```

```python
df.to_csv('../Data/NaverMoivesRankingList.csv', encoding='utf-8', index=False)
```
