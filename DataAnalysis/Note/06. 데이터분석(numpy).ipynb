### DataFrame Merging

```python
df1 = pd.DataFrame(
    {
        'key': list('bbacaab'),
        'data1': range(7)
    }
)
df1
```

```python
df2 = pd.DataFrame(
    {
        'key': list('abd'),
        'data2': range(3)
    }
)
df2
```

```python
# default : Inner Join
pd.merge(df1,df2, on='key')
```

```python
# Outter Join
pd.merge(df1,df2, on='key', how='outer')
```

```python
# df1 기준으로 합치기
pd.merge(df1,df2, on='key', how='left')
```

```python
# df2 기준으로 합치기
pd.merge(df1,df2, on='key', how='right')
# data1에는 d가 없어서 NaN이 출력됨
```

```python
# different Key name
df1 = pd.DataFrame(
    {
        'lkey' : list('bbacab'),
        'data1' : range(6)
    }
)
df1
```

```python
df2 = pd.DataFrame(
    {
        'rkey' : list('ababd'),
        'data2' : range(5)
    }
)
df2
```

```python
pd.merge(df1, df2, left_on='lkey', right_on='rkey', how='inner')
```

### DataFrame의 Concatenation

```python
df1 = pd.DataFrame(
    np.arange(6).reshape(3,2),
    index=['a','b','c'],
    columns=['one','two']
)
df2 = pd.DataFrame(
    5 + np.arange(4).reshape(2,2),
    index=['a','c'],
    columns=['three','four']
)
```

```python
df1
```

```python
df2
```

```python
# df1 과 df2 합치기 (merge 아님, 개는 key값을 따지고 얘는 걍 합치는거 )

pd.concat([df1,df2], axis=1)
```

```python
pd.concat([df1,df2], axis=0)
```

```python
pd.concat([df1,df2], axis=1, ignore_index=True)
```

```python
pd.concat([df1,df2], axis=0, ignore_index=True)
```

### Category

```python
df3 = pd.DataFrame(
    {
        'id' : [1,2,3,4,5,6],
        'raw_grade' : ['a','b','b','a','a','e']
    }
)
df3
```

```python
# Category 자료형으로 변환하기
df3['grade'] = df3['raw_grade'].astype('category')
df3
```

```python
df3.raw_grade
```

```python
df3.grade
```

```python
# category를 이용하여 자료 변형
df3.grade.cat.categories = ['very good','good','very bad']
df3
```

```python
df3.sort_values(by='grade')
```

```python
df3.grade.info()
```

### Data의 범위 정하기

```python
ages=[20,22,25,27,21,23,37,31,61,45,41,32]
bins=[18,25,35,60,100]
```

```python
cats = pd.cut(ages, bins)
cats
```

```python
# 범주관련 코드 보기
cats.codes
```

```python
# 범주관련 빈도수 보기
cats.value_counts()
```

```python
# caategory 이름 정하기
group_names = ['Youth','YoungAdult','MiddleAged','Senior']
cat2 = pd.cut(ages, bins, labels=group_names)
cat2
```

```python
cat2.value_counts()
```
