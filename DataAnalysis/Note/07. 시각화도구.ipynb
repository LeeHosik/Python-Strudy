### 전출지별이 서울특별시, 전입지별이 서울특별시가 아닌경우 의 Data를 보고 싶다 
 : 서울에서 다른 지역으로 이동한 Data를 추출

```python
# df_seoul = 
dfTest = df.copy()
```

```python
df_seoul = dfTest[(dfTest['전출지별'] == '서울특별시') & (dfTest['전입지별'] != '서울특별시')]
# df_seoul
```

```python
# df_seoul은 전출지별 column이 다 서울에서 나가는거라 의미가 없음으로 삭제

df_seoul.drop('전출지별',axis='columns', inplace=True)
# df_seoul.drop(columns='전출지별', inplace=True)#?
```

```python
df_seoul.head(2)
```

```python
# df_seoul 전입지별 -> '전입지' 로 컬럼 이름 변경
df_seoul.rename(columns={'전입지별':'전입지'}, inplace=True)
#  = df_seoul.rename({'전입지별':'전입지'}, axis=1, inplace=True)
df_seoul.head(2)
```

```python
# 전입지가 전국, 부산광역시, 댁구광역시 인것만 출력하기(단, | 사용금지)
list= ['전국','부산광역시','대구광역시']
df_seoul = df_seoul['전입지'].isin(list)
df_seoul[list]
```

```python
# df_seoul의 index를 전입지로 변경 
df_seoul.set_index('전입지', inplace=True)
```

```python
df_seoul.head()
```

```python
# 서울에서 경기도로 이동한 인구 데이터 값만 선택
sr_one = df_seoul.loc['경기도']
sr_one.head()
```

```python
import platform
from matplotlib import font_manager, rc
```

```python
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
plt.plot(
    sr_one.index,
    sr_one.values
    )
plt.show()
```

```python
type(sr_one) #= Series
```

```python
# 챠트 제목 추가
plt.plot(sr_one)
plt.title('서울 => 경기 인구 이동')

# 축 이름
plt.xlabel('연도')
plt.ylabel('이동 인구수')

plt.show()
```

### 챠트 꾸미기

```python
# 챠트 사이즈 지정(가로 14인치, 세로 5인치)
plt.figure(figsize=(14,5))

# x축 눈금 라벨 회전하기
# plt.xticks(rotation=45)
plt.xticks(rotation=90)
# plt.xticks(rotation=135)

# y축 눈금 라벨 회전하기
plt.yticks(rotation=45)

# 챠트 제목 추가
plt.plot(sr_one)
plt.title('서울 => 경기 인구 이동')

# 축 이름
plt.xlabel('연도')
plt.ylabel('이동 인구수')

# 범례 표시
plt.legend(labels = ['서울->경기'], loc = 'best')

plt.show()
```
