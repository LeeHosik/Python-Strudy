15보다 적게, 30보다 크게, 그 사이 로 3분류

```python
# 2분위수와 3분위수의 근사 정수치를 선택 
train['LowFare'] = train['Fare'] <15
train['MediumFare'] = (train['Fare'] >= 15) & (train['Fare'] < 30)
train['HighFare'] = train['Fare'] >= 30

train[['Fare','LowFare','MediumFare','HighFare']].head(10)
```

```python
trainTest = train.copy()
```

```python
trainTest.loc[train['LowFare'] == True , 'FareType'] = 'Low'
trainTest.loc[train['MediumFare'] == True , 'FareType'] = 'Medium'
trainTest.loc[train['HighFare'] == True , 'FareType'] = 'High'
```

```python
train[['Fare','FareType']].head(10)
```

```python
# Python 의 함수를 이용하여 처리하기

def check_fare_type(fare):
    if fare < 15 :
        return 'Low'
    elif fare>= 15 and fare <30:
        return 'Medium'
    else : return 'High'
```

```python
trainTest['FareType2'] = trainTest['Fare'].apply(check_fare_type)
trainTest[['Fare','FareType','FareType2']]
```
