### 지도에 마커 표시하기
- 서울 시내 주요 대학교의 위치를 표시하기

```python
df = pd.read_excel('../Data/서울지역 대학교 위치.xlsx')
df.columns = ['학교명','위도','경도']
df
```

```python
# Map 만들기
seoul_map2 = folium.Map(
    location=[37.55, 126.98],
    tiles="Stamen Terrain",
    zoom_start=12 # zomm
)

# 대학교위치정보를 Marker로 표시
for name, lat, lng in zip(df.학교명, df.위도, df.경도):
    folium.Marker([lat, lng], popup=name).add_to(seoul_map2)

seoul_map2
```

---
### 지도에 원형 마커 표시하기

```python
# Map 만들기
seoul_map2 = folium.Map(
    location=[37.55, 126.98],
    tiles="Stamen Terrain",
    zoom_start=12 # zomm
)

# 대학교위치정보를 Marker로 표시
for name, lat, lng in zip(df.학교명, df.위도, df.경도):
    folium.Marker([lat, lng], popup=name).add_to(seoul_map2)

# 대학교 위치 정보를 CirvleMarker로 표시하기
for name, lat, lng in zip(df.학교명, df.위도, df.경도):
    folium.CircleMarker([lat,lng],
                        radius=10, # 원의 반지름
                        color='brown', # 원 둘레의 색상
                        fill=True,
                        fill_color='coral', # 원을 채우는 색
                        fill_opacity=0.7 # 투명도
                        ).add_to(seoul_map2)

seoul_map2
```

# 지도 영역에 단계구분포(Choropleth Map)표시
- 행정구역과 같이 지도상의 어떤 경계에 둘어싸인 영역에 색을 칠하거나 음영등을 칠하는 방법

```python
df = pd.read_excel('../Data/경기도인구데이터.xlsx', index_col='구분')
df.head()
```

```python
df.columns = df.columns.map(str)
df.columns
```

```python
import json
```

```python
geo_path = '../Data/경기도행정구역경계.json'
geo_data = json.load(open(geo_path, encoding='utf-8'))
geo_data
```

```python
# Map 만들기
g_map = folium.Map(
    location=[37.55, 126.98],
    tiles="Stamen Terrain",
    zoom_start=9 # zomm
)

# 출력할 연도 선택
year = '2007'

# choropleth class로 단계 구분 표시
folium.Choropleth(
    geo_data=geo_data, # 지도 검색
    data=df[year], # 표시하려는 데이터
    columns=[df.index, df[year]], # 열 지정
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.3,
    threshold_scale=[10000, 100000, 300000, 500000, 700000], # 한계점
    key_on='feature.properties.name'
).add_to(g_map)

# 지도를 HTML 파일로 저장하기
g_map.save('../Data/gyeongido_population_'+year+'.html')
g_map
```
