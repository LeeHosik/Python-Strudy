```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
import folium
warnings.filterwarnings("ignore")
```

```python
# 한글 폰트 문제 해결 
# matplotlib은 한글 폰트를 지원하지 않음
# os정보
import platform

# font_manager : 폰트 관리 모듈
# rc : 폰트 변경 모듈
from matplotlib import font_manager, rc
# unicode 설정
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
df = pd.read_csv('../Data/UpgradePaiksCoffee.csv')
df.info()
```

```python
# df['lat'] = pd.to_numeric(df['lat'])
df[['lng','lat']] = df[['lng','lat']].apply(pd.to_numeric, errors = 'coerce')
```

```python
# Map 만들기
PaiksCoffee = folium.Map(
    # location = [37.55, 126.98], 
    tiles = 'Stamen Terrain', 
    # tiles = 'Stamen Toner', 
    zoom_start = 12
    )

# # 대학교 위치정보를 Marker로 표시|


# 대학교 위치정보를 CircleMarker로 표시
# for name, lat,lng in zip(df.학교명, df.위도, df.경도):
#     folium.CircleMarker([lat, lng], radius=10; color='brown'; fill=True; fill_color='coral'; fill_opacity=0.7; popup = name ).add_to(seoul_map2)

for name, lat, lng in zip(df.ShopName, df.lat, df.lng):
    try:
        folium.CircleMarker([lat, lng],
                            radius = 10,
                            color = 'brown',
                            fill = True,
                            fill_color = 'coral',
                            fill_opacity = 0.7,
                            popup = name).add_to(PaiksCoffee)
    except:print(f'{name} error')
    


PaiksCoffee
```
