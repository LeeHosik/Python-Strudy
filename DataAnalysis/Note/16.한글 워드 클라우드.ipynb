```python
import nltk
from konlpy.corpus import kolaw #< 여기 헌법이 들어있다 함 
from konlpy.tag import Okt
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# 한글 폰트 문제 해결 
# matplotlib은 한글 폰트를 지원하지 않음
# os정보
import platform

# font_manager : 폰트 관리 모듈
# rc : 폰트 변경 모듈
from matplotlib import font_manager, rc
# unicode 설정
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
t = Okt()
```

```python
ko_conn_text = kolaw.open('constitution.txt').read()
ko_conn_text
```

```python
token_ko = t.nouns(ko_conn_text)
token_ko
```

```python
# 불용어 추가
stop_words = ['제','월','일','그','이','바','및','의','기타','입','정','수','관','때','위','경우','모든','안','인','장','거나','다만']
token_ko = [each_word for each_word in token_ko if each_word not in stop_words]
```

```python
# 상위 50개 단어 확인
ko = nltk.Text(token_ko)
ko.vocab().most_common(50)
```

```python
# 상위 50개의 빈도수 챠트 만들기
plt.figure(figsize=(12,6))
ko.plot(50)
plt.show()
```

```python
# 상위 500개만 wordcloud로 그리기
data = ko.vocab().most_common(500)
tmp_data = dict(data)
wordcloud = WordCloud(
    backgound_color = 'white',
    font_path = 'AppleGothic',# //\ window는 필요 >> \ 'c:/Windows/Fonts/malgun.ttf'
    relative_scaling = 0.3
).generate_from_frequencies(tmp_data)
```

```python
plt.figure(figsize=(16, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
