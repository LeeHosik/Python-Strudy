### Pandas의 내장 그래픽 도구

```python
# 남북한 발전량 데이터셋 불러오기
data = pd.read_excel('../Data/남북한발전전력량.xlsx', engine='openpyxl') # python version이 구형이면 안불러와질때가 있다고 하는데 그때 사용하는녀석
data 
```

```python
# 남한, 북한 발전량 합계 데이터만 추출하여 df_ns 로 만들기
# df_ns의 행 index를 South 와 North로 변경하기
df_ns = data.iloc[[0,5],2:]
df_ns.index = ['South','North']
df_ns
```

```python
# 그래프를 그릴껀데 내용들이 다 숫자값이냐 확인
df_ns.info()
```

```python
# 죄다 object네 
# object -> int로 변환

df_ns = df_ns.astype('int')
df_ns.info()
```

```python
df_ns.head()
```

```python
# 선 그래프 그리기 전에 column들도 숫자인지 문자인지 확인해보자
df_ns.columns #   dtype='object' 네 , 해보고 문제 생기면 애도 숫자로 바꿔야 겠따 라고 생각하고 
```

```python
# 선그래프
df_ns.plot()
```
